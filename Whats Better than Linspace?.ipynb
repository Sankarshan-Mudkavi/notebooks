# Effective determination of interpolants with upper bounds error.

From what I've experience, it is usually bad idea to use a large number of interpolants at equally spaced intervals. Computer nowadays have tones of ram and considering the nuances of computing isn't too much of a concern anymore. However, I've learned that if we don't start caring about efficiency, we never will. 

Consider some spaced out [interpolants.](http://en.wikipedia.org/wiki/Interpolation)

$$  \alpha < x_1 < x_2 < \cdots < x_n < \beta $$. What we would like to so if apply a function $f$ onto these points to essentially get a grasp of what the function looks like, among other properties.

**How do we pick the best spacing?**

Naively, and with a lot of processing power, we usually just space a lot of points evenly. 

>Lets make an array from $[\alpha, \beta]$ and let the subdivisions to go infinity... Lets just pick 10000 or something I guess.

No. Don't do that.

Today I'm going to show you two types of ways you can figure out how many points to need. One will be a static method that uses our knowledge of the Taylor Remainder Theorem, and the other is an adaptive method that recursively finds the best divisions that take into account the linear behavior of some sections in the data. 

## Static Piecewise Linear Interpolation

Consider the following situation where $y_i = f(x_i)$ for some function $f(x)$ where the $x_i$ are evenly spaced.

```python
"""Set up a piecewise linear interpolant 
   of sin(2πx) on a uniform, nine-point 
   partition of [0, 1]"""

x = linspace(0, 1, 9)
x2 = linspace(0, 1, 100)
ylim(-1.5,1.5)
xlim(-.2, 1.2)
title("A piecewise linear function \n with $9$ linearly spaced interpolants")
xlabel("x")
ylabel("y")
plot(x,np.sin(2*np.pi*x), 'o-k')
plot(x2,np.sin(2*np.pi*x2), '-r')
show()
```

This method has some problems. 

1. The function above is fairly tame, and equally setting up intervals does not lose a lot of the information. But what about a more complex function?
2. While more intervals give us better answers, is throwing more RAM at a problem really the best solution?
3. We are using a series of piecewise linear functions to approximate the function, can we get an upper bound on the error? 

Consider the example below.

```python
"""Set up a piecewise linear interpolant 
   of the matlab humps(x) function on a 
   uniform, nine-point partition of [0, 3]"""

f = lambda x: 1/((x - 0.3)**2 + 0.01) + 1/((x-0.9)**2 + 0.04) - 6

x = linspace(0, 3, 9)
x2 = linspace(0, 3, 400)
xlim(-.2, 3.2)
title("A tricky function with \n9 linear interpolants")
xlabel("$x$")
ylabel("$f(x)$")
plot(x2,f(x2), '-r', lw=1)
plot(x,f(x), 'o-k')
show()
```

Look at that error! 

As mentioned before computers nowadays can generate a `linspace(0,100, 100000)` just fine, it doesn't mean that we have to. Instead we should be having conversations about the error and how deciding the step value instead of just making it arbitrarily large. 

### Some Math.

What we are essentially doing is saying that on each interval on the $x$ axis is approximating that function as a line $L(x)$ 

$$ f(c) = L(c) + \text{Error}$$

For any point $c \in [x_I, x_{I+1}]$ we can say that $f(c)$ can be approximated as that line, which is the value of our line at $c$ plus some error term.

With the assistance of the Taylor Polynomials, we can easily quantify that error.

### Remember Taylor Polynomials???

From calculus we remember that any* function can be expanded as a power series. For a general function $f(x)$ expanded about the point $x=a$:

$$f(x) = f(a) + f'(a)(x-a) + \frac{f''(a)}{2}(x-a)^2 + \frac{f'''(a)}{3!}(x-a)^2 + \cdots$$ 

Notice that the first two terms actually correspond to a linear approximation, something we've all learned in first and second year calculus.

$$f(x) = \overbrace{f(a) + f'(a)(x-a)}^{L(x)} + \frac{f''(a)}{2}(x-a)^2 + \cdots $$ 

If you *really* remember, we can also say that by Taylor's Remainder Theorem, 

$$ |F(c) - L(z)| \le \frac{M_2 h^2}{8}$$

Where $M_2$ is the upper bounds the $|f''(x)|$ over the interval $[\alpha, \beta]$ and $h$ is width of the subdivision. Since we are using our trusty linspace function, we also know that location of each interval is equal to:

$$ x_i = \alpha + \frac{i-1}{n-1}(\beta - \alpha) \Rightarrow h = x_{i+1} - x_i = \left( \frac{\beta-\alpha}{n-1}\right) $$

To guarantee that the error between $f$ and our approximation $l$ is less than some tolerance level, say, $\epsilon$, we make the statement that

$$ |f(c) - L(c)| \le \frac{M_2}{8}\left( \frac{\beta-\alpha}{n-1}\right)^2 \le \epsilon $$

From the above statement we can directly see that $n$ must satisfy the condition: 

$$ n \ge 1 + (\beta - \alpha)\sqrt{M_2 / 8 \epsilon}$$

So, Lets implement this and see how it goes.

```python
def static_interpolation(fn, lower, upper, M2, tolerance):
    """Static Interpolation of a linear piecewise function
    
    Attributes:
        fn (function) - function we are interpolating
        M2 (float)    - upper bound on |f''(x)| over the lower and upper bound
        lower (float) - lower bount
        upper (float) - upper bound
        tolerance (float) - maximun acceptable error
    
    Returns:
        x (numpy.array) - x values
        y (numpy.array) - y values
        n (int) - number of divisions
        
    """
    
    n = np.ceil(1 + (upper- lower) * np.sqrt(M2 / (8 * tolerance)))
    x = np.linspace(lower, upper, n)    return x, fn(x), n
```

```python

M2 = np.max(np.gradient(f(np.linspace(0, 3, 100)))) #totally cheating
x, y, n = static_interpolation(f, 0., 3., M2, 0.005)
x2 = linspace(0, 3, 4000)
xlim(-.2, 3.2)
title("Static-Interpolation")
xlabel("$x$")
ylabel("$f(x)$")
plot(x, y, 'o-k', label="f(x) for n={}".format(n))
plot(x2,f(x2), '-r', label="f(x)")
legend()
print("Hey! We only needed {} intervals! No guess work.".format(n))
```

This is great! We see that to only have an error of $0.001$ we only need 146 intervals. We don't need to just throw in a million 'to be safe', we just need to figure out what we need, and then get what we need.

There is one more problem however, notice that after the big hump we have *a lot* of intervals that really do nothing. The peak has a lot of sparse information and the rest of the intervals didn't need to be there.

What if this was a really expensive calculation? We would have wasted quite a bit of time! *We can do better.*

Instead of using a static method of calculating the idea interval, we can instead us an adaptive method. That finds those areas that might have a lot of flat boring parts and thins down the interpolants.

## Adaptive Piecewise Linear Interpolation

To address the problem, we can use a recursive partitioning algorithm that discovers where $f$ is linear and where it is not and cluster the breakpoints effectively.

So what defines a good breakpoint? We say that the interval is appropriate if 

$$ \left| f \left( \frac{xL + xR}{2} \right) - \frac{f(xL) + f(xR)}{2} \right| \le \epsilon \text{ or } xR - xL \le h_{min}$$

That this distribes if the different between the midpoint and the two bounds of that sub interval.

```python
def adaptive_interpolation(fn, lower, upper, hmin, tolerance):
    """Adaptive Interpolation of a linear piecewise function
    
    Attributes:
        fn (function) - function we are interpolatin
        lower (float) - lower bound
        upper (float) - upper bound
        hmin (float) - lower bound on interval spaces
        tolerance (float) - maximun acceptable error
    
    Returns:
        x (numpy.array) - x values
        y (numpy.array) - y values
        
    """
    fL, fU = fn(lower), fn(upper)
    if (upper - lower) <= hmin:
        x = np.array([lower, upper])
        y = np.array([fL, fU])
        return x, y, shape(x)[0]
    else:
        mid_x = np.mean([lower, upper])
        mid_fx = np.mean([fL, fU])
        f_mid_x = fn(mid_x)    
        if np.abs(mid_fx - f_mid_x) <= tolerance:        
            x = np.array([lower, upper])
            y = np.array([fL, fU])
            return x, y, shape(x)[0]
        else:
            xl, yl, _ = adaptive_interpolation(fn, lower, mid_x, hmin, tolerance)
            xr, yr, _ = adaptive_interpolation(fn, mid_x, upper, hmin, tolerance)
            x = np.concatenate((xl, xr[1:])) 
            y = np.concatenate((yl, yr[1:]))
            return x, y, shape(x)[0]
```

What we should notice is that it should not just pick a global equally spaved interval. Instead I should realise that the flat runs require less points while the bumps should have more resolution.

```python
figure(2)
subplot(212)
ax, ay, n = adaptive_interpolation(f, lower=0.0, upper=3.0, hmin=.05, tolerance=0.1)
xlim(-.2, 3.2)
title("Adaptive-Interpolation")
plot(ax, ay, 'o-k', label="f(x) for n={}".format(n))
plot(x2,f(x2), '-r', label="f(x)")
text(1.6,25, "Notice how many points we've saved!", fontsize=14)
legend()

figure(1)
subplot(211)
M2 = np.max(np.gradient(f(np.linspace(0, 3, 100)))) #totally cheating
x, y, n = static_interpolation(f, 0., 3., M2, 0.005)
xlim(-.2, 3.2)
title("Static-Interpolation")
plot(x, y, 'o-k', label="f(x) for n={}".format(n))
plot(x2,f(x2), '-r', label="f(x)")
legend()
```

We see here that by using half the number of divisions, we can get pretty much the same resolution on the two bumps while ignoring the flat long 'runs' after $x=1.5$ If we really want to go crazy, we can set our $h_{min}$ to be very small.
