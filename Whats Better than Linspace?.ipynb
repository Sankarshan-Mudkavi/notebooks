# Interesting investigation of picking breakpoints.

It is usually a bad idea to use a large number of interpolants at equally spaced intervals. Especially if it is over a large interval $[\alpha, \beta]$.

$$  \alpha < x_1 < x_2 < \cdots < x_n < \beta $$

#### How do we decide the best number of points?

Well, there are both static and adaptive methods. The idea is that we can plot the points $[f(x_1), f(x_2), \cdots, f(x_n)]$. versus our set of $x_i$ points. 

## Piecewise Linear Interpolation

Consider the following situation where $y_i = f(x_i)$ for some function $f(x)$

```python
"""Set up a piecewise linear interpolant 
   of sin(2πx) on a uniform, nine-point 
   partition of [0, 1]"""

x = linspace(-0, 1, 9)
ylim(-1.5,1.5)
xlim(-.2, 1.2)
title("A piecewise linear function \n with $9$ linearly spaced interpolants")
xlabel("x")
ylabel("y")
plot(x,np.sin(2*np.pi*x), 'o-k')
show()
```

This inherently has some problems. The function above is fairly tame, and equally setting up intervals does not loose a lot of the information. But how about a more complex function?

```python
"""Set up a piecewise linear interpolant 
   of sin(2πx) on a uniform, nine-point 
   partition of [0, 1]"""

f = lambda x: 1/((x - 0.3)**2 + 0.01) + 1/((x-0.9)**2 + 0.04) - 6

x = linspace(0, 3, 9)
x2 = linspace(0, 3, 400)
xlim(-.2, 3.2)
title("A trickier piecewise linear function")
xlabel("$x$")
ylabel("$f(x)$")
plot(x2,f(x2), '-r', lw=1)
plot(x,f(x), '^-k')
show()
```

While computers nowadays can generate a `linspace(0,100, 10000000)` just fine, it doesn't mean that we have to. If we consider each line segment to be a function $L$ we can see that 

$$ f(c) = L(c) + \frac{f''(c)}{2} (c-x_i)(c-x_{i+1}) $$

This approximation is saying that for a point $c \in [x_i, x_{i+1}]$ we can interpolate that as $L(c)$ which is the value of our line at $c$ plus the a bit of the second derivative. 

This is because of the properties of taylor polynomials.

### Remember Taylor Polynomals???

$$f(x) = f(a) + f'(a)(x-a) + \frac{f''(a)}{2}(x-a)^2 + \frac{f'''(a)}{3!}(x-a)^2 + \cdots$$ 

But notice that our line segment is just the first two terms. This linear approximation about $c$ is something we've all learned in first year calculus. Save for some minor modifications where our one sided $(x-a)^2$ becomes $(c-x_i)(c-x_{i+1})$

$$f(x) = \overbrace{f(a) + f'(a)(x-a)}^{L(x)} + \frac{f''(a)}{2}(x-a)^2 + R(x)$$ 

If you *really* remember, we can also say that by Taylor's Remainder Theorem, 

$$ |F(c) - L(z)| \le \frac{M_2 h^2}{8}$$

For some $M_2$ that bounds the second derivative of $f$ and $h$ that is width of the subdivision. Moreover, since we are using linespace, we also know that the width is equal to the following:

$$ x_i = \alpha + \frac{i-1}{n-1}(\beta - \alpha)$$

To basically ganruntee that the error from $f$ and our approximation $l$ is less than some tolerance level $\epsilon$ we can say that 

$$ |f(c) - L(c)| \le \frac{M_2 h^2}{8}\left( \frac{\beta-\alpha}{n-1}\right) \le \epsilon $$

From this we can see that $n$ must satisfy the condition 

$$ n \ge 1 + (\beta - \alpha)\sqrt{M_2 / 8 \epsilon}$$

```python
def static_interpolation(fn, lower, upper, M2, tolerance):
    """Static Interpolation of a linear piecewise function
    
    Attributes:
        fn (function) - function we are interpolating
        M2 (float)    - upper bound on |f''(x)| over the lower and upper bound
        lower (float) - lower bount
        upper (float) - upper bound
        tolerance (float) - maximun acceptable error
    
    Returns:
        x (numpy.array) - x values
        y (numpy.array) - y values
        n (int) - number of divisions
        
    """
    
    n = np.ceil(1 + (upper- lower) * np.sqrt(M2 / (8 * tolerance)))
    x = np.linspace(lower, upper, n)
    return x, fn(x), n
```

```python
f = lambda x: 1/((x - 0.3)**2 + 0.01) + 1/((x-0.9)**2 + 0.04) - 6
M2 = np.max(np.gradient(f(np.linspace(0, 2, 100)))) #totally cheating

x, y, n = static_interpolation(f, 0., 2., M2, 0.001)

x2 = linspace(0, 3, 100)
xlim(-.2, 3.2)
title("Application of Static-Interpolation")
xlabel("$x$")
ylabel("$f(x)$")
plot(x2,f(x2), '-r', lw=2)
plot(x, y, '.--k')
print("Hey! We only needed {} intervals! No guess work.".format(n))
```

However, notice that after 0.5, the main hump we have $a lot$ of intervals that really do nothing. What if this was a really expensive calculation. We would have wasted quite a bit of time!

We can do better. 

Instead of using a static method of calculating the idea interval, we can instead us an adaptive method.

### Adaptive Piecewise Linear Interpolation

To address the problem, we use a recursive partitioning algorithm that discovers where $f$ is linear and where it is not and clusters the break points effectively.
