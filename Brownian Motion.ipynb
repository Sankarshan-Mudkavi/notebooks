## Two main ideas

1. There is a timestep $\tau$ that has the two following properties
 
 * That is **short** enough such that : $ f(x,t) \approx f(x, t+\tau)$
       
 * That is **long** enough that millions of collisions occur between $[t, t+\tau]$

We now that the mean time for collisons is $10^{-10}s$. There is no way to describe these collisions explicitly.

2. There is no hope to describe these solvent-pollen explicitly so he does so with a probaility distribution.

    $$ \Phi (\epsilon) d \epsilon = \text{Probability that the pollen grain is net displaced by }\tau$$

## What are some properties of $\Phi(\epsilon)$?

**It is a probability density**

* It is normalizable 

$$ \int_{-\infty}^{\infty} \Phi (\epsilon) d\epsilon = 1 $$
       
**Collisions are unbiased**

* It is symmetric 

$$ \Phi (\epsilon) = \Phi (-\epsilon) $$
    
**Over the small timestep**

* Sharply peaked at the origin!

$$f(x,t) \approx f(x, t + \tau)$$

## How does the density change over a step size $\tau$?

Aside: If the pollen grains are independent. $f(x,y) = N_{total} \cdot p(x,t)$

$$ f(x, t + \tau) = \int_{-\infty}^{\infty} f(x - \epsilon, t)\Phi(\epsilon)d \epsilon$$

This function in a way 'counts' all the ways you can move to position $x$ over $[t, t+\tau]$. What Einstien was looking at was a conditional probability...

There are Analogous to the [Chapman-Kolomagrov Equations](http://en.wikipedia.org/wiki/Chapman%E2%80%93Kolmogorov_equation). These are also considered the master equations since it provides a method to generate all moments of the distribution.

To make headway, Einstein made a taylor expansion of both sides since $\tau, \epsilon$ is small...

$$ f(x, t+\tau) \approx f(x,t) + \tau \frac{d}{dt} f(x,t)$$
$$ f(x-\epsilon, t) \approx f(x,t) -\epsilon \frac{d}{dx} f(x,t) + \frac{\epsilon^2}{2} \frac{d^2}{dx^2}\cdots$$

## Now we sub it all in back into the integral

$$ f(x,t) +  \tau \frac{d}{dt} f(x,t) =  \int_{-\infty}^{\infty} f(x,t) -\epsilon \frac{d}{dx} f(x,t) + \frac{\epsilon^2}{2} \frac{d^2}{dx^2}\cdots \Phi(\epsilon)d \epsilon $$

$$ f(x,t) +  \tau \frac{d}{dt} f(x,t) =   f(x,t)\int_{-\infty}^{\infty}\Phi(\epsilon)d \epsilon - \frac{d}{dx} f(x,t)\int_{-\infty}^{\infty}\epsilon \Phi(\epsilon)d \epsilon + \frac{d^2}{dx^2}\int_{-\infty}^{\infty}\frac{\epsilon^2}{2} \Phi(\epsilon)d \epsilon $$

$$ f(x,t) +  \tau \frac{d}{dt} f(x,t) =   f(x,t)\underbrace{\int_{-\infty}^{\infty}\Phi(\epsilon)d \epsilon}_{1} - \frac{d}{dx} f(x,t)\underbrace{\int_{-\infty}^{\infty}\epsilon \Phi(\epsilon)d \epsilon}_{0} + \frac{d^2}{dx^2}\underbrace{\int_{-\infty}^{\infty}\frac{\epsilon^2}{2} \Phi(\epsilon)d \epsilon}_{Variance} $$


All of this goes to show that if we consider the variance can be some constant $D$,

$$ \frac{df}{dt} = D \frac{d^2f}{dx^2} \Rightarrow f(x,t) = \frac{1}{\sqrt{4\pi Dt}} e^{\frac{-x^2}{4Dt}},\ \ t>0$$

Therefore we can see that $\mu_x = 0, \sigma_x^2 = 2Dt$

### Using Statistical Physics (Osmotic Pressure and Gibbs Free Energy)

$$ D = \frac{K_B T}{6 \pi \nu R}$$

- $K_B$ - Boltzmann Constant  
- $T$ - Tempurature
- $\nu$ - Viscosity
- $R$ - Radius of the Pollen

# The Experiment

Einstien Proposes the following experiment.

### Measure the Root-Mean-Square Displacement

$$ \sigma_x = \sqrt{2D t} = \sqrt{\frac{K_B T}{3 \pi \nu R}} t^{1/2}$$

Therefore, the slope of $\sigma_x$ versus $t^{1/2} \propto \sqrt{K_B}$. This was the **first time** ever irreproductbility was used as a reproducible measure!

This experiment won [Perrin the Nobel Prize](http://www.nobelprize.org/nobel_prizes/physics/laureates/1926/perrin-lecture.html) in 1926.
