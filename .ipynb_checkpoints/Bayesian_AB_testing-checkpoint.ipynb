# Baysian AB testing conversion rates

* how can we use the Beta-Binomial model in A/B Tests?

```python
%pylab inline
figsize(12,8)
```

```python
from scipy.stats import beta
```

```python
dataA = {
    "total":100.,
    "conversions":30
}

dataB = {
    "total":100.,
    "conversions":35.
}
```

With this data we could naively say that 

     conversion_rate_A = 0.30
     conversion_rate_b = 0.35

Let us assume a flat prior for our conversion rates and plot the priors.

```python
def plot_posteriors(x ,data, label):
    # the addition of 1 is to assume the flat prior
    y = beta.pdf(x, 1+data['conversions'], 1 + data['total'] - data['conversions'])
    plt.plot(x, y, label=label)
    plt.legend()
    return
```

```python
x = np.linspace(0.1, 0.6, 200)

plot_posteriors(x, dataA, "A")
plot_posteriors(x, dataB, "B")
```

```python
def sample_posteriors(dataA, dataB):
    samplesA = beta.rvs(1 + dataA['conversions'],
                        1 + dataA['total'] - dataA['conversions'],
                        size = 10000)
    samplesB = beta.rvs(1 + dataB['conversions'],
                        1 + dataB['total'] - dataB['conversions'],
                        size = 10000)
    return samplesA, samplesB
    
```

```python
sA, sB = sample_posteriors(dataA, dataB)
```

What we did here was sample 10000 times the distribution. Now we are doing to ask ourselves what the probability of sampleA being greater than sampleB

```python
(sA>sB).mean()
```

Given the data you might say that B is better than A but with a 22% change of being wrong...

Lets create more data.

```python
dataA = {
    "total":10000.,
    "conversions":35
}

dataB = {
    "total":10000.,
    "conversions":55.
}
```

```python
x = np.linspace(0.0, 0.01, 400)

plot_posteriors(x, dataA, 'A')
plot_posteriors(x, dataB, 'B')
```

```python
sA, sB = sample_posteriors(dataA, dataB)
```

```python
(sA>sB).mean()
```

This is the baysian $p$-value. Which states that if you said sampleB > sampleA, you would be wrong, 1.9% of the time.
